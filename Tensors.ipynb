The index_select method allows you to extract parts of a tensor along a given dimension. 

Note that the method takes three arguments: the tensor to operate on, 

the dimension along which to extract data, and **the tensor containing the indices**. 

(In our case, index = torch.LongTensor([0, 1]) is the tensor of indices)

In the following, we construct a 3x3 tensor, and then extract data along each of the two dimensions.

```python
a = torch.FloatTensor([[1 ,2, 3],[4, 5, 6], [7, 8, 9]])
a
```

```python
index = torch.LongTensor([0, 1])
b = torch.index_select(a, 0, index)
b
```

```python
c = torch.index_select(a, 1, index)
c
```

The masked_select method , illustrated in the following Listing, 

allows you to select elements given a Boolean mask.

```python
a = torch.FloatTensor([[1 ,2, 3],[4, 5, 6], [7, 8, 9]])
a
```

```python
mask = torch.ByteTensor([[0, 1, 0],[1, 1, 1],[0, 1, 0]])
mask
```

```python
b = torch.masked_select(a, mask.bool())
b
```

The addbmm function (where bmm stands for batch matrix-matrix product) 
<br>allows you to perform the computation p * m + q * [a1 * b1 + a2 * b2 + ...], 
<br>where p and q are scalars, and m, a1, b1, a2, and b2 are tensors. 
<br>Note that the addbmm function takes parameters p and q with default values 
<br>equal to one and that tensors such as a1 and a2 are provided by stacking them along the first dimension. 

```python
a = torch.ones(2, 2, 3)
a
```

```python
b = torch.ones(2, 3, 2)
b
```

```python
m = torch.ones(2,2)
m
```

```python
torch.addbmm(2, m, 3, a, b)
```

```python
torch.addbmm(1, m, 1, a, b)
```

```python
torch.addbmm(m, a, b)
```

The addmm function is a non-batch version of addbmm 
<br>that allows you to perform the computation p * m + q * a * b, 
<br>where p and q are scalars, and m, a, and b are tensors. 
<br>Note that the addmm function takes parameters p and q with default values equal to one.

```python
a = torch.ones(2, 3)
a
```

```python
b = torch.ones(3, 2)
b
```

```python
m = torch.ones(2,2)
m
```

```python
torch.addmm(m, a, b)
```

```python
torch.addmm(2, m, 3, a, b)
```

```python
torch.addmm(1, m, 1, a, b)
```
